```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np 

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# Read the Mouse and Drug Data and the Clinical Trial Data
df_mouse = pd.read_csv("data/mouse_drug_data.csv")
df_clinical = pd.read_csv("data/clinicaltrial_data.csv")

# Combine the data into a single dataset
df = pd.merge(df_mouse, df_clinical, how = "outer")

# Display the data table for preview
df.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
df_group = df.groupby(["Drug","Timepoint"])

# Convert to DataFrame
df_tumor = pd.DataFrame(df_group.mean()["Tumor Volume (mm3)"]).reset_index()

# Preview DataFrame
df_tumor.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
# Convert to DataFrame
df_error = pd.DataFrame(df_group.sem()["Tumor Volume (mm3)"]).reset_index()
# Preview DataFrame
df_error.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
tumor_table = df_tumor.set_index(["Timepoint"]).pivot(columns = "Drug", values = "Tumor Volume (mm3)")

# Preview that Reformatting worked
tumor_table.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
error_table = df_error.set_index(["Timepoint"]).pivot(columns = "Drug", values = "Tumor Volume (mm3)")

# Preview that Reformatting worked
error_table.head()
```

```python
plt.errorbar(tumor_table.index, tumor_table["Capomulin"],
             yerr=error_table["Capomulin"], fmt="o", ls = ":", lw = 1, ms = 4,
             color="red", ecolor = "grey", capsize = 3, label="Capomulin")

plt.errorbar(tumor_table.index, tumor_table["Infubinol"],
             yerr=error_table["Infubinol"], fmt="^", ls = ":", lw = 1, ms = 4,
             color="blue", ecolor = "grey", capsize = 3, label="Infubinol")

plt.errorbar(tumor_table.index, tumor_table["Ketapril"],
             yerr=error_table["Ketapril"], fmt="s", ls = ":", lw = 1, ms = 4,
             color="green", ecolor = "grey", capsize = 3, label="Ketapril")

plt.errorbar(tumor_table.index, tumor_table["Placebo"],
             yerr=error_table["Placebo"], fmt="D", ls = ":", lw = 1, ms = 4,
             color="black", ecolor = "grey", capsize = 3, label="Placebo")

plt.title("Tumor Response to Treatment")
plt.xlabel("Time (Days)")
plt.ylabel("Tumor Volume(mm3)")
plt.legend (loc="best", fontsize="small", fancybox=True)

plt.xlim(-5,50)
plt.ylim(30,75)
plt.grid(True)

plt.savefig("Tumor Response Plot.png")
plt.show()
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
mean_meta = df_group['Metastatic Sites'].mean()

# Convert to DataFrame
df_mean = pd.DataFrame(mean_meta).reset_index()

# Preview DataFrame
df_mean.head()
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
sem_meta = df_group['Metastatic Sites'].sem()

# Convert to DataFrame
df_sem = pd.DataFrame(sem_meta).reset_index()

# Preview DataFrame
df_sem.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
mean_table = df_mean.set_index(["Timepoint"]).pivot(columns = "Drug", values = "Metastatic Sites")

# Preview that Reformatting worked
mean_table.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
sem_table = df_sem.set_index(["Timepoint"]).pivot(columns = "Drug", values = "Metastatic Sites")

# Preview that Reformatting worked
sem_table.head()
```

```python
plt.errorbar(mean_table.index, mean_table["Capomulin"],
             yerr=sem_table["Capomulin"], fmt="o", ls = ":", lw = 1, ms = 4,
             color="red", ecolor = "grey", capsize = 3, label="Capomulin")

plt.errorbar(mean_table.index, mean_table["Infubinol"],
             yerr=sem_table["Infubinol"], fmt="^", ls = ":", lw = 1, ms = 4,
             color="blue", ecolor = "grey", capsize = 3, label="Infubinol")

plt.errorbar(mean_table.index, mean_table["Ketapril"],
             yerr=sem_table["Ketapril"], fmt="s", ls = ":", lw = 1, ms = 4,
             color="green", ecolor = "grey", capsize = 3, label="Ketapril")

plt.errorbar(mean_table.index, mean_table["Placebo"],
             yerr=sem_table["Placebo"], fmt="D", ls = ":", lw = 1, ms = 4,
             color="black", ecolor = "grey", capsize = 3, label="Placebo")

plt.title("Metastatic Spread During Treatment")
plt.xlabel("Treatment Duration (Days)")
plt.ylabel("Metastatic Sites")
plt.legend (loc="best", fontsize="small", fancybox=True)

plt.xlim(-5,50)
plt.ylim(-0.5,4)
plt.grid(True)

plt.savefig("Metastatic Spread Plot.png")
plt.show()
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
df_mice = df_group["Mouse ID"].count().reset_index().rename(columns = {"Mouse ID": "Mouse Count"})

# Preview DataFrame
df_mice.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
mouse_table = df_mice.set_index(["Timepoint"]).pivot(columns = "Drug", values = "Mouse Count")

# Preview that Reformatting worked
mouse_table.head()
```

```python
mouse_table = 100*(mouse_table/mouse_table.iloc[0,:])
mouse_table.head()
```

```python
plt.plot(mouse_table.index, mouse_table["Capomulin"], marker="o", ls = ":", lw = 1, ms = 4,
             color="red", label="Capomulin")

plt.plot(mouse_table.index, mouse_table["Infubinol"], marker="^", ls = ":", lw = 1, ms = 4,
             color="blue", label="Infubinol")

plt.plot(mouse_table.index, mouse_table["Ketapril"], marker="s", ls = ":", lw = 1, ms = 4,
             color="green", label="Ketapril")

plt.plot(mouse_table.index, mouse_table["Placebo"], marker="D", ls = ":", lw = 1, ms = 4,
             color="black", label="Placebo")

plt.title("Survival During Treatment")
plt.xlabel("Time (Days)")
plt.ylabel("Survival Rate (%)")
plt.legend (loc="best", fontsize="small", fancybox=True)

plt.xlim(-2.5,47.5)
plt.ylim(30,105)
plt.grid(True)

plt.savefig("Metastatic Spread Plot.png")
plt.show()
```
